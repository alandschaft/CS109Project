## Test generation routines

* A single test squence is a list of test actions
* We focus on testing term selection, so term selection is the single test action we define
* "SELECT_TERM" test action has a random parameter - value between 0 and 1. This value indicated the position in the list of the selected term. For example if we have 10 terms in the list and the random number is 0.4 then the term in position 4 is selected.

```python
class TEST_ACTION:
    SELECT_TERM = 1    

random.seed(datetime.now())    

def generate_random_test_sequence(number_of_acions = 3):
    actions = []
    for t in range(0, number_of_acions):
        actions.append(get_random_action())
    return actions

def get_random_action():
    test_type = TEST_ACTION.SELECT_TERM
    select_position = round(random.uniform(0.0, 1.0), 2)
    return test_type, select_position
```

## Test target parameter - Cosine Similarity

One of the things we would like to test is how "similar" are the clinical trials in the list that is presented to the user at the end of the test sequence, to the list of terms that were selected during the test sequence. We are using tf-idf based cosine similarity - a common method in information retrieval ( https://en.wikipedia.org/wiki/Cosine_similarity ) to calculate the level of similarity between the list of terms and the list of documents.

```python
def get_average_cosine_similarity(docs, terms):
    if not docs or not terms:
        return 0
    # Create tf-idf vectorizer and matrix from the documents table
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(docs)
    # Vectorize the list of terms
    tfidf_terms = tfidf_vectorizer.transform([' '.join(terms)])
    # Calculate cosine similarities between the list of terms and each document
    score = cosine_similarity(tfidf_terms, tfidf_matrix)
    # return average score
    return np.mean(score)

def extract_relevance(result):
    if len(result['ui_docs']) < 0:
        return None, None
    # for each test result we calculate the tf-idf based average cosine similarity of the presented documents to the list of selected terms
    docs_relevance=get_average_cosine_similarity(
        [' '.join(doc) for doc in [doc['terms'] for doc in result['ui_docs']]]
        , [t['text'] for t in result['selected_terms']])
    return docs_relevance
```

## Execution of a test sequence and aggregation of target parametes

The following functions execute a single test sequence and aggregate the target test parameters that we collect from each test sequence.

```python
def execute_select_term(session, select_position):
    # select the term in the select_position position in the list of terms presented in the ui
    if len(session['ui_terms']) <= 0: 
        return None
    term_index =int((len(session['ui_terms'])*select_position-1))
    selected_term=session['ui_terms'][term_index]['text']
    return on_select_term(session, selected_term)

test_mapper = {
    TEST_ACTION.SELECT_TERM: execute_select_term}

def execute_test_sequence(test_sequence):
    session = create_session_data()
    # session id is not important for testing - generate just a random number
    session['session_id'] = randint(1,9999999)
    n_docs_begin = len(session['candidate_docs'])
    result = None
    # execute each action in the sequence of test actions
    for action in test_sequence:
        result=test_mapper[action[0]](session, action[1])
    if result == None:
        return None, None
    # extract relevance score
    relevance_score = extract_relevance(result)
    # extract average reduction of clicking on a tag
    n_docs_end = len(session['candidate_docs'])
    n_select_tag = [i[0] for i in test_sequence].count(TEST_ACTION.SELECT_TERM)
    reduction_score = (n_docs_begin - n_docs_end) / n_select_tag
    return relevance_score, reduction_score
```

## Run a sequence of tests

```python
def run_test(list_of_tests, k_terms, m_terms):
    # set up dictionary entry where we will collect the data for this test
    res={}
    res['relevance score'] = []
    res['reduction score'] = []
    # set ap the parameters in our term selection algorithm
    search_lib.k_terms=k_terms
    search_lib.m_terms=m_terms
    for test in list_of_tests:
            relevance_score, reduction_score = execute_test_sequence(test)
            if relevance_score == None:
                continue
            else:
                res['relevance score'].append(relevance_score)
                res['reduction score'].append(reduction_score)
    return res



def _run_test_scheme(n_test_sequences=20, k_max=10):
    list_of_tests = [generate_random_test_sequence() for _ in xrange(n_test_sequences)]
    dict_results={}
    # For each (k,m) combination - run the same set of tests
    for k in range(2,k_max):
        for m in range(1,k):
            res=run_test(list_of_tests=list_of_tests, k_terms=k, m_terms=m)
            dict_results[(k,m)]=res
    return dict_results

```
